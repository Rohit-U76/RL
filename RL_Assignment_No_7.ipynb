```python
import gymnasium as gym
import numpy as np
import time

# -------------------------------
# 1️⃣ Create deterministic FrozenLake
# -------------------------------
env = gym.make('FrozenLake-v1', map_name="4x4", is_slippery=False, render_mode='human')
n_states = env.observation_space.n
n_actions = env.action_space.n

print(f"Action space size: {n_actions}")
print(f"State space size: {n_states}")

# -------------------------------
# 2️⃣ Initialize policy and value function
# -------------------------------
pi = np.zeros(n_states, dtype=int)  # initial policy: always LEFT
V = np.zeros(n_states)              # initial value function
gamma = 0.99                        # discount factor
theta = 1e-6                         # convergence threshold

# Transition probabilities
P = env.unwrapped.P  # P[s][a] = list of (prob, next_state, reward, done)

# -------------------------------
# 3️⃣ Policy Iteration
# -------------------------------
policy_stable = False

while not policy_stable:
    # -------- Policy Evaluation --------
    while True:
        delta = 0
        for s in range(n_states):
            v = V[s]
            V[s] = 0
            for a in range(n_actions):
                for prob, s_next, reward, done in P[s][a]:
                    action_prob = 1.0 if a == pi[s] else 0.0
                    V[s] += action_prob * prob * (reward + gamma * V[s_next])
            delta = max(delta, abs(v - V[s]))
        if delta < theta:
            break

    # -------- Policy Improvement --------
    policy_stable = True
    for s in range(n_states):
        old_action = pi[s]
        Q = np.zeros(n_actions)
        for a in range(n_actions):
            for prob, s_next, reward, done in P[s][a]:
                Q[a] += prob * (reward + gamma * V[s_next])
        pi[s] = np.argmax(Q)
        if old_action != pi[s]:
            policy_stable = False

# -------------------------------
# 4️⃣ Display Results
# -------------------------------
print("\nOptimal Policy (0=LEFT,1=DOWN,2=RIGHT,3=UP):")
print(pi.reshape(4,4))

print("\nOptimal State-Value Function:")
print(np.round(V.reshape(4,4), 3))

# -------------------------------
# 5️⃣ Run an episode using the optimal policy
# -------------------------------
state, _ = env.reset()
done = False
step = 0

print("\nRunning episode using the optimal policy...")
while not done and step < 20:
    env.render()
    action = pi[state]
    state, reward, done, info, _ = env.step(action)
    step += 1
    time.sleep(1)

env.render()
if reward == 1:
    print(f"\nReached the Goal in {step} steps!")
else:
    print(f"\nFell into a hole!")

env.close()
```

Assignment 7: Value Iteration & Policy Iteration – FrozenLake

Objective: Implement value iteration and policy iteration to find the optimal value function and optimal policy for the deterministic FrozenLake environment.

Environment Reference: FrozenLake – Gymnasium

1. Introduction

FrozenLake Environment: 4×4 grid

Start (S), Goal (G), Frozen (F), Holes (H)

Actions: 0 = LEFT, 1 = DOWN, 2 = RIGHT, 3 = UP

Rewards: 1 at the goal, 0 elsewhere

Deterministic Environment: is_slippery=False

We implement:

Value Iteration: Compute optimal state-values → extract policy

Policy Iteration: Iteratively evaluate and improve policy → convergence
