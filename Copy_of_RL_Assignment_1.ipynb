```python
# Practical Assignment 1
# Basic Reinforcement Learning using Reward-Based Learning
# Multi-Armed Bandit Example

import random

# Actions (study methods)
actions = ["Reading", "Practice Problems", "Videos"]

# Initialize action values (estimated rewards)
action_value = [0.0, 0.0, 0.0]

# Count how many times each action is selected
action_count = [0, 0, 0]

# Number of learning episodes
episodes = 100


def get_reward(action):
    """
    Simulated reward from environment.
    Practice Problems gives better performance.
    """

    if action == 1:  # Practice Problems
        return random.choice([1, 1, 1, 0])  # Higher chance of reward

    elif action == 2:  # Videos
        return random.choice([1, 0])  # Moderate chance

    else:  # Reading
        return random.choice([0, 0, 1])  # Lower chance


# Reinforcement Learning loop (Trial-and-Error learning)
for episode in range(episodes):

    # Select action randomly (exploration)
    action = random.randint(0, 2)

    # Get reward from environment
    reward = get_reward(action)

    # Update action count
    action_count[action] += 1

    # Update estimated reward using average reward formula
    action_value[action] += (reward - action_value[action]) / action_count[action]


# Display results
print("Estimated reward for each study method:\n")

for i in range(len(actions)):
    print(f"{actions[i]} : {action_value[i]:.2f}")

# Identify best action
best_action = actions[action_value.index(max(action_value))]

print("\nBest study method learned:", best_action)
```

## 3. Problem Description

A student tries different study methods before exams:

1. Reading textbooks
2. Solving practice problems
3. Watching educational videos

After each attempt (**action**), the student receives performance feedback (**reward**):

* **Reward = 1** → Good performance
* **Reward = 0** → Poor performance

The goal of the **agent** is to learn which study method gives the **highest average reward over time**.

* The learner (student)

### 4.2 Environment

* Exam performance evaluation system

| 1         | Practice Problems |
| 2         | Watching Videos   |

### 4.4 Rewards

| Outcome          | Reward |
| ---------------- | ------ |
| Good Performance | 1      |
| Poor Performance | 0      |

---

## 5. Algorithm Used

### Basic Reinforcement Learning (Multi-Armed Bandit)

This algorithm:

* Selects an action randomly
* Receives reward from the environment
* Updates the estimated value of the selected action using the **average reward**
* Repeats the process for multiple episodes

No **Q-table**, **Bellman equation**, or **discount factor** is used.

---

## 6. Program Execution Steps

1. Initialize action values to zero

2. Initialize action count to zero

3. Repeat for a fixed number of episodes

   **Steps inside each episode:**

   * Select an action randomly
   * Obtain reward from the environment
   * Update action value using the **average reward formula**

4. Display the learned values

5. Identify the best action
